Bag-of-Words modeli bir metni kelime frekanslarna gre temsil eden basit bir modeldir. Bu modelde metin iindeki her kelimenin ka kez getii saylr ve bu bilgiler bir vektr olarak saklanr. BoW modeli kelimelerin srasn dikkate almaz, sadece kelime sklklarna odaklanr.



TF-IDF (Term Frequency-Inverse Document Frequency), bir kelimenin bir belgede ne kadar nemli olduunu belirleyen bir metin temsil yntemidir. TF, kelimenin bir belgede ne sklkta getiini ifade ederken, IDF, kelimenin tm belgeler arasnda ne kadar nadir olduunu ler. Bu sayede sk geen yaygn kelimelere daha dk, nadir ancak belirleyici olan kelimelere daha yksek arlk verir.



```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer

# Metin dosyasndan veri okuma
def derlemden_veri_okuma(dosya_adi):
    with open(dosya_adi, "r", encoding="utf-8") as file:
        veri = file.readlines()
    # lk 10.000 cmleyi kullanma (hzl almas iin)
    veri = [satir.strip() for satir in veri[:10000]]  # 10,000 rnek cmleyle snrlandrld
    return veri

# Dosya adn belirtin (rnein "derlem.txt")
dosya_adi = "dosyalar//odtu_derlemi.txt"

# Derlemden veri okuma
veri = derlemden_veri_okuma(dosya_adi)

# Bag-of-Words Modeli
vectorizer_bow = CountVectorizer()
X_bow = vectorizer_bow.fit_transform(veri)

# BoW vektrlerini gsterme
print("Bag-of-Words Vektrleri:")
print(pd.DataFrame(X_bow.toarray(), columns=vectorizer_bow.get_feature_names_out()).head())

# TF-IDF Modeli
vectorizer_tfidf = TfidfVectorizer()
X_tfidf = vectorizer_tfidf.fit_transform(veri)

# TF-IDF vektrlerini gsterme
print("\nTF-IDF Vektrleri:")
print(pd.DataFrame(X_tfidf.toarray(), columns=vectorizer_tfidf.get_feature_names_out()).head())
```

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Kullancdan salanan cmle
yeni_cumle = ["st aa yamur"]

# Yeni cmleyi vektrletirme
yeni_cumle_vect = vectorizer_bow.transform(yeni_cumle)

# Yeni cmle ile tm cmleler arasndaki benzerlikleri hesaplama
benzerlikler = cosine_similarity(yeni_cumle_vect, X_bow).flatten()

# En benzer 10 cmleyi bulma
en_benzer_indices = benzerlikler.argsort()[-10:][::-1]  # En yksek benzerlikten balayarak sralar

# En benzer cmleleri ve benzerliklerini gsterme
print(f"'{yeni_cumle[0]}' cmlesine en yakn 10 cmle (BoW ile):")
for idx in en_benzer_indices:
    if benzerlikler[idx] > 0:
        print(f"Cmle: '{veri[idx]}' - Benzerlik: {benzerlikler[idx]:.2f}")
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from TurkishStemmer import TurkishStemmer
import re

# TurkishStemmer kullanarak gvdeleme
stemmer = TurkishStemmer()

# Noktalama iaretlerinden temizleme ve kk harfe evirme fonksiyonu
def temizle(metin):
    metin = metin.lower()  # Metni kk harfe evirir
    metin = re.sub(r'[^\w\s]', '', metin)  # Noktalama iaretlerini kaldrr
    return metin

# Gvdeleme fonksiyonu
def turkce_govdeleme(metin):
    kelimeler = metin.split()
    govdeler = [stemmer.stem(kelime) for kelime in kelimeler]
    return ' '.join(govdeler)

# Metin dosyasndan veri okuma, temizleme ve gvdeleme uygulama
def derlemden_veri_okuma(dosya_adi):
    with open(dosya_adi, "r", encoding="utf-8") as file:
        veri = file.readlines()
    # lk 10.000 cmleyi kullanma (hzl almas iin)
    orijinal_veri = [temizle(satir.strip()) for satir in veri[:10000]]  # Temizlenmi orijinal metinler
    govdeli_veri = [turkce_govdeleme(satir) for satir in orijinal_veri]  # Gvdeleme uyguland
    return orijinal_veri, govdeli_veri

# Dosya adn belirtin (rnein "odtu_derlemi.txt")
dosya_adi = "dosyalar//odtu_derlemi.txt"

# Derlemden veri okuma, temizleme ve gvdeleme uygulama
orijinal_veri, govdeli_veri = derlemden_veri_okuma(dosya_adi)

# TF-IDF Modeli
vectorizer_tfidf = TfidfVectorizer()
X_tfidf = vectorizer_tfidf.fit_transform(govdeli_veri)

# Bag-of-Words Modeli
vectorizer_bow = CountVectorizer()
X_bow = vectorizer_bow.fit_transform(govdeli_veri)

# Kullancdan salanan cmle
yeni_cumle = ["st aa yamur"]

# Yeni cmleyi temizleme ve gvdeleme uygulama
yeni_cumle_temiz = temizle(yeni_cumle[0])
yeni_cumle_govdeli = [turkce_govdeleme(yeni_cumle_temiz)]

# Yeni cmleyi TF-IDF ile vektrletirme
yeni_cumle_tfidf = vectorizer_tfidf.transform(yeni_cumle_govdeli)

# Yeni cmleyi Bag-of-Words ile vektrletirme
yeni_cumle_bow = vectorizer_bow.transform(yeni_cumle_govdeli)

# Yeni cmle ile tm cmleler arasndaki benzerlikleri hesaplama (TF-IDF)
benzerlikler_tfidf = cosine_similarity(yeni_cumle_tfidf, X_tfidf).flatten()

# Yeni cmle ile tm cmleler arasndaki benzerlikleri hesaplama (BoW)
benzerlikler_bow = cosine_similarity(yeni_cumle_bow, X_bow).flatten()

# En benzer 10 cmleyi bulma (TF-IDF)
en_benzer_indices_tfidf = benzerlikler_tfidf.argsort()[-10:][::-1]  # En yksek benzerlikten balayarak sralar

# En benzer 10 cmleyi bulma (BoW)
en_benzer_indices_bow = benzerlikler_bow.argsort()[-10:][::-1]  # En yksek benzerlikten balayarak sralar

# En benzer cmleleri ve benzerliklerini orijinal metinlerle gsterme (BoW)
print(f"'{yeni_cumle[0]}' cmlesine en yakn 10 cmle (BoW ve TurkishStemmer ile Gvdeleme):")
for idx in en_benzer_indices_bow:
    if benzerlikler_bow[idx] > 0:
        print(f"Cmle: '{orijinal_veri[idx]}' - Benzerlik: {benzerlikler_bow[idx]:.2f}")
        
print("\n" + "="*50 + "\n")
        
# En benzer cmleleri ve benzerliklerini orijinal metinlerle gsterme (TF-IDF)
print(f"'{yeni_cumle[0]}' cmlesine en yakn 10 cmle (TF-IDF ve TurkishStemmer ile Gvdeleme):")
for idx in en_benzer_indices_tfidf:
    if benzerlikler_tfidf[idx] > 0:
        print(f"Cmle: '{orijinal_veri[idx]}' - Benzerlik: {benzerlikler_tfidf[idx]:.2f}")
```

Information Retrieval

```python
import os
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from TurkishStemmer import TurkishStemmer
import re

# TurkishStemmer kullanarak gvdeleme
stemmer = TurkishStemmer()

# Metni temizleme fonksiyonu (lowercase ve noktalama iaretlerinden arndrma)
def temizle(metin):
    metin = metin.lower()  # Kk harfe evirme
    metin = re.sub(r'[^\w\s]', '', metin)  # Noktalama iaretlerini kaldrma
    return metin

# Gvdeleme fonksiyonu
def turkce_govdeleme(metin):
    kelimeler = metin.split()
    govdeler = [stemmer.stem(kelime) for kelime in kelimeler]
    return ' '.join(govdeler)

# Klasrdeki metin dosyalarn oku ve temizle
def dosyalari_oku_ve_temizle(klasor_adi):
    dosyalar = os.listdir(klasor_adi)
    metinler = []
    dosya_adlari = []
    for dosya in dosyalar:
        dosya_yolu = os.path.join(klasor_adi, dosya)
        with open(dosya_yolu, "r", encoding="utf-8") as f:
            metin = f.read()
            temiz_metin = temizle(metin)
            govdeli_metin = turkce_govdeleme(temiz_metin)
            metinler.append(govdeli_metin)
            dosya_adlari.append(dosya)
    return metinler, dosya_adlari

# Klasr ad
klasor_adi = "dosyalar//haberler"

# Klasrdeki metin dosyalarn oku ve temizle
metinler, dosya_adlari = dosyalari_oku_ve_temizle(klasor_adi)

# TF-IDF ile metin dosyalarn vektrletirme
vectorizer_tfidf = TfidfVectorizer()
X_tfidf = vectorizer_tfidf.fit_transform(metinler)

# Kullancdan arama sorgusu al
arama_sorgusu = input("Arama yapmak istediiniz cmleyi girin: ")

# Arama sorgusunu temizle ve gvdelenmi hale getir
arama_sorgusu_temiz = temizle(arama_sorgusu)
arama_sorgusu_govdeli = [turkce_govdeleme(arama_sorgusu_temiz)]

# Arama sorgusunu TF-IDF ile vektrletirme
arama_sorgusu_vect = vectorizer_tfidf.transform(arama_sorgusu_govdeli)

# Arama sorgusu ile dosyalar arasndaki benzerlikleri hesaplama
benzerlikler = cosine_similarity(arama_sorgusu_vect, X_tfidf).flatten()

# Benzerliklere gre en iyi eleen dosyalar bulma ve sralama
en_benzer_indices = benzerlikler.argsort()[-5:][::-1]  # En yksek benzerlikten balayarak sralar

# En benzer dosyalar ve benzerlik oranlarn gsterme
print(f"'{arama_sorgusu}' arama sorgusuna en yakn dosyalar:")
for idx in en_benzer_indices:
    if benzerlikler[idx] > 0:
        print(f"Dosya: '{dosya_adlari[idx]}' - Benzerlik: {benzerlikler[idx]:.2f}")
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import numpy as np

# Trke durdurma kelimeleri listesi
turkish_stop_words = [
    've', 'bu', 'da', 'de', 'bir', 'ile', 'iin', 'ise', 'olan', 'gibi', 'en', 'btn', 'olarak', 'diye', 'ben', 'sen',
    'nn', 'nin', 'n', 'in', 'byk', 'nasl', 'bile', 'baka', 'nce', 'deil', 'u', 'var', 'ya', 'gre',
    'ok', 'veya', 'fakat', 'ancak', 'nk', 'eer', 'kadar', 'her', 'tm', 'ama', 'ne', 'daha', 'sonra', 'ki'
]

# Derlemi dosyadan okuma
with open('dosyalar//odtu_derlemi.txt', 'r', encoding='utf-8') as file:
    corpus = file.readlines()

# BoW modeli ile kelime sayma (stop words filtreleme)
vectorizer = CountVectorizer(stop_words=turkish_stop_words)
X = vectorizer.fit_transform(corpus)

# Kelimelerin ve frekanslarnn alnmas
word_freq = np.asarray(X.sum(axis=0)).flatten()  # Bellek dostu sklk hesaplamas
word_freq_dict = dict(zip(vectorizer.get_feature_names_out(), word_freq))

# Kelime bulutu oluturma
wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(word_freq_dict)

# Kelime bulutunu grselletirme
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
